```python
analysis_radius = float(CONFIG['Track Generation']['BoundingSphereRadius'])
velocity_bins = np.linspace(1e5,7e5,5)

alphas = pint.generate_alphas(velocity_bins, theta_bin_n=5, phi_bin_n=4, radius=analysis_radius)
```

```python
i = 0 # track 1
time = out[i]['time']-out[i]['time'][0]
time = np.concatenate([time, time[-1] + np.linspace(0,5000,51, dtype=int)])

strt_indx_len = randrange(0, 60)
strt_accel_pad = np.zeros([strt_indx_len, 3])

end_indx_len = randrange(0, 60)
end_accel_pad = np.zeros([end_indx_len, 3])

accels = []
sensors_pos = []

cnt = 0
for key in out[i]:
    if key not in ['time', 'particle_location']:
        print(key)
        accels_sensor = [[0] * 3 for l in range(len(out[i]['time']))]
        for tms in range(0,len(out[i]['time'])):
            for dim in range(0,3):
                accels_sensor[tms][dim] = accels_w_noise[tms + (len(out[track-1]['time']) * cnt)][dim]
        accels.append(np.concatenate([strt_accel_pad,accels_sensor,end_accel_pad]))
        sensors_pos.append(DETECTOR_CHARACTERISTICS[key]['position'])
        cnt += 1
```
